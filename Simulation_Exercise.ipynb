# **Simulation Exercise**

**Step 1 -** *Let $t \sim$ exp($\mu$). Find the probability $P(t \geq 2\mu)$*

**Step 2 -**
* Find the probability $P(t \geq 10)$, for $\mu$ $\in$ $\{1,...,100\}$
* Find the probability $P(t \geq 15)$, for $\mu$ $\in$ $\{1,...,100\}$
* Find the probability $P(t \geq 20)$, for $\mu$ $\in$ $\{1,...,100\}$

**Step 3 -** *Create the DataFrame and Graph*

```python
# Step 1

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

mu = float(input("Enter a value for mu: "))
n = 0 # runs
c = 0 # counter

while n < 10000:
  t = np.random.exponential(mu)
  if t >= 2 * mu: c = c + 1
  n = n + 1
print('probability: ', c / n)

# Step 2

values = range (1, 101) # probably range
p1 = [] # P(t>=10) values
p2 = [] # P(t>=15) values
p3 = [] # P(t>=20) values

for mu in values:
  n = 0 #runs
  c = [0, 0, 0] #counter
  while n <= 1000:
    t = np.random.exponential(mu)
    if t >= 10: c[0] = c[0] + 1
    if t >= 15: c[1] = c[1] + 1
    if t >= 20: c[2] = c[2] + 1
    n = n + 1
  p1.append(c[0] / n)
  p2.append(c[1] / n)
  p3.append(c[2] / n)

# Step 3

results = pd.DataFrame({'mu values': values, 'P(t > 10)': p1, 'P(t > 15)': p2, 'P(t > 20)': p3}, index = values) # creating the dataframe
plt.plot(results['mu values'], results['P(t > 10)'], c = 'r') # plotting P(t>=10) values
plt.plot(results['mu values'], results['P(t > 15)'], c = 'b') # plotting P(t>=15) values
plt.plot(results['mu values'], results['P(t > 20)'], c = 'g') # plotting P(t>=20) values
plt.ylabel('probabilities') # labeling y axis
plt.xlabel('values') # labeling x axis
plt.show() # showing visualization
display(results) # showing the data
```
