```python
# Step 1: Importing Libraries
import pandas as pd
import seaborn as sns

# Step 2: Loading the Iris Dataset
iris = sns.load_dataset('iris')

# Step 3: Dropping the "species" column
iris.drop(labels = 'species', axis = 1, inplace = True)

# Step 4: Add "sepal_sum" and "petal sum" Columns to the Data Frame
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']

# Step 5: Create a New Data Frame for Mean, Standard Deviation, Minimum, and Maximum Values
mean = list(iris.mean())
std = list(iris.std())
min = list(iris.min())
max = list(iris.max())

summary_dict = { 'mean': mean,
                 'std' : std,
                 'min' : min,
                 'max' : max }

summary_df = pd.DataFrame(data = summary_dict)

# The final product
summary_df
```
